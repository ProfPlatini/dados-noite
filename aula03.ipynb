````python
quantidade_produtos = [3,5,6,10]
preco_produtos = [10,5,7,100]
nomes_produtos = ["Caçulinha","Pastel","Refrigerante","Combos"]

array_produtos = np.array(quantidade_produtos)
array_precos = np.array(preco_produtos)

faturamento = array_produtos * array_precos


for nome,preco in zip(nomes_produtos,faturamento):
    print(f"{nome} : {preco}")
````

### Exercício de fixação 
- Criar uma lista que irá possuir 10 nomes de produtos.
- Criar uma lista que irá possuir o preço de custo (atacado).
- Criar uma lista que irá possuir o preço de venda (varejo).
- Ao final, exibir o lucro obtido em cada produto.

````python
#Lista de nomes
produtos = ["Celular","Televisão","Notebook","Computador","Teclado"]
#Lista de preços atacado
preco_atacado = [2000,3000,4500,5000,100]
#Lista de preços varejo
preco_varejo = [2500,4000,5000,7000,120]
#Array
array_atacado = np.array(preco_atacado)
array_varejo = np.array(preco_varejo)

lucro = array_varejo - array_atacado
lucro_obtido = np.sum(lucro)
print(f"O lucro obtido foi R${lucro_obtido}")

for nome,lucro_produto in zip(produtos,lucro):
    print(f"O lucro do produto {nome} foi {lucro_produto}")
````
