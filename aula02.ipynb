Programa que avalia as notas que o usuário do nosso aplicativo inseriu.

```python
avaliacao = float(input("Digite a sua avaliação: 0 - 10 estrelas"))

if avaliacao > 8:
    print("Nota excelente!")
elif avaliacao > 6:
    print("Avaliação regular")
elif avaliacao > 4:
    print("Avaliação ruim")
else:
    print("Avaliação péssima")
```

Sistema que irá verificar se o filme avaliado é bom ou não.
- Criar uma variável que irá armazenar o nome do filme 
- Criar uma variável que irá armazenar a nota do filme (0 - 5)
- Condição: Se possuir 5 estrelas, filme Vencedor do Oscar. 
- Se possuir de 3 a 4,99, filme Bom para assistir
- Abaixo disso, filme não indicado.

```python
nome_filme = input("Digite o nome do filme")
nota_filme = float(input("Digite a nota do filme: "))
if nota_filme == 5:
    print(f"FILME {nome_filme} VENCEDOR DO OSCAR")
elif nota_filme >= 3:
    print(f"FILME {nome_filme} BOM PARA ASSISTIR")
else:
    print(f"FILME {nome_filme} NÃO INDICADO!")
```

```python
nome_filme = input("Digite o nome do filme")
nota_filme = float(input("Digite a nota do filme: "))
if nota_filme <= 0 or nota_filme > 5 :
    print("❌ Nota inválida")
else:
    if nota_filme == 5:
        print(f"FILME {nome_filme} VENCEDOR DO OSCAR")
    elif nota_filme >= 3:
        print(f"FILME {nome_filme} BOM PARA ASSISTIR")
    else:
        print(f"FILME {nome_filme} NÃO INDICADO!")
    
```
